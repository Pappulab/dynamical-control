```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import networkx as nx

condensatelist=['Cyto_Centrosome','Cyto_Neuronal Granule','Cyto_P body','Cyto_Spindle Apparatus','Cyto_Stress Granule','Nuc_Cajal Body','Nuc_DNA Damage Foci','Nuc_Gemini of Cajal Body','Nuc_Histone Locus Body','Nuc_Nuclear Speckle','Nuc_Nucleolus','Nuc_Paraspeckle','Nuc_PcG Body','Nuc_PML Nuclear Body','Nuc_Sam68 Nuclear Body']

listcyto = [i for i in condensatelist if i.startswith('Cyto_')]
numcyto=len(listcyto)

listnuc = [i for i in condensatelist if i.startswith('Nuc_')]
numnuc=len(listnuc)

colormapnodes=[]
# Create colormap for cytoplamic bodies
n_lines = numcyto+1
x = np.linspace(0, 10, 100)
y = np.sin(x[:, None] + np.pi * np.linspace(0, 1, n_lines))
c = np.arange(1, n_lines + 1)
norm = mpl.colors.Normalize(vmin=c.min(), vmax=c.max())
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.Blues)
cmap.set_array([])
for i in range(0,numcyto):
    tcolor=cmap.to_rgba(i+2)
    colormapnodes.append(mpl.colors.to_hex(tcolor))
print(colormapnodes)

# Create colormap for nuclear bodies
n_lines2 = numnuc+1
x2 = np.linspace(0, 10, 100)
y2 = np.sin(x2[:, None] + np.pi * np.linspace(0, 1, n_lines2))
c2 = np.arange(1, n_lines2 + 1)
norm = mpl.colors.Normalize(vmin=c2.min(), vmax=c2.max())
cmap = mpl.cm.ScalarMappable(norm=norm, cmap=mpl.cm.Reds)
cmap.set_array([])
for i in range(0,numnuc):
    tcolor=cmap.to_rgba(i+2)
    colormapnodes.append(mpl.colors.to_hex(tcolor))
print(colormapnodes)

colormapnodes2=[]
for c in colormapnodes:
    colormapnodes2.append(c[1:])
    
print(colormapnodes2)


def combine_hex_values(d):
    d_items = sorted(d.items())
    tot_weight = sum(d.values())
    red = int(sum([int(k[:2], 16)*v for k, v in d_items])/tot_weight)
    green = int(sum([int(k[2:4], 16)*v for k, v in d_items])/tot_weight)
    blue = int(sum([int(k[4:6], 16)*v for k, v in d_items])/tot_weight)
    zpad = lambda x: x if len(x)==2 else '0' + x
    return zpad(hex(red)[2:]) + zpad(hex(green)[2:]) + zpad(hex(blue)[2:])


allLists=[]
for c in condensatelist:
    # Load in composition data
    df=pd.read_excel('DrLLPS_compositions.xlsx', sheet_name=c,usecols='F')
    currgenelist=df['Gene Name'].tolist()
    
    cleanedList = [x for x in currgenelist if str(x) != 'nan']
    allLists.append(cleanedList)

numcomp=[]
colormapedges=[]
G = nx.Graph()
for l1 in range(0,len(allLists)):
    G.add_node(condensatelist[l1])
    numcomp.append(5*len(allLists[l1]))
    for l2 in range(l1+1,len(allLists)):
        tmp=list(set(allLists[l1]) & set(allLists[l2]))
        #print(tmp)
        #print(len(tmp))
        #hi
        G.add_edge(condensatelist[l1],condensatelist[l2],weight=len(tmp)*0.025)
        if len(tmp)==0:
            print(condensatelist[l1]+' with '+condensatelist[l2])
        tmpcolor=combine_hex_values({colormapnodes2[l1]: 1.0, colormapnodes2[l2]: 1.0})
        colormapedges.append('#'+tmpcolor)
        
        
edges = G.edges()
weights = [G[u][v]['weight'] for u,v in edges]
pos = nx.circular_layout(G)

#nx.draw(G, pos, edges=edges, edge_color=colors, width=weights)
        
fig=plt.subplots(figsize=(20,10))            
nx.draw_networkx(G,pos,node_size=numcomp,node_color=colormapnodes,edge_color=colormapedges,with_labels=True,width=weights)
#plt.show() 

plt.rcParams['pdf.fonttype'] = 42 # Makes text editiable
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.sans-serif'] = "Arial"
plt.rcParams['font.family'] = "sans-serif"
#plt.savefig('DrLLPS_Condensate_Components_Graph.pdf', dpi=300, transparent=True, bbox_inches='tight')
```
