# Plotting Tie-Lines

## Defining System Names

```python
SYS_NAMES = ["8_3_HetAB_X_X", "8_3_HetAB_HomA_X", "8_3_HetAB_HomA_HomB"]
```

```python
SYS_TITLES = {"8_3_HetAB_X_X" : "Heterotypic Only",
             "8_3_HetAB_HomA_X" : "Heterotypic + $RNA$\nHomotypic",
             "8_3_HetAB_HomA_HomB" : "Heterotypic + $RNA$\nHomotypic + Whi3 Homotypic"}
```

```python
erKW = {'capsize':2.0, 'capthick':2.0, 'elinewidth':1.0}
```

## Function Of A Line

```python
def pl_line(m, b, x):
    return m*x + b
```

## Plotting For Fig S9

### V2

```python
myFig, myAx = plt.subplots(2, 3, figsize=(12*0.85,8*0.85), sharex=False, sharey=False, constrained_layout=True)

for sysID, sysName in enumerate(SYS_NAMES):
    myAx[0, sysID].set_title(SYS_TITLES[sysName])
    for rowID in range(2):

        CoexistenceData = np.loadtxt(f"{sysName}_Filtered.txt")
        c0_1, c0_2, cden_1, cden_2, dcden_1, dcden_2, csol_1, csol_2, dcsol_1, dcsol_2 = CoexistenceData.T
        myAx[rowID, sysID].errorbar(y=cden_1, x=cden_2, yerr=dcden_1, xerr=dcden_2,  
                     color='#374099', fmt='.', alpha=0.8, label=r'Dense Phase', **erKW, zorder=0)
        myAx[rowID, sysID].errorbar(y=csol_1, x=csol_2, yerr=dcsol_1, xerr=dcsol_2,
                         color='#b62326',  fmt='.', alpha=0.8, label=r'Dilute Phase', **erKW, zorder=0)
        
        TieLineData = np.loadtxt(f"{sysName}_TieLineData.txt")
        c0_1, c0_2 = TieLineData.T[:2,:]
        myAx[rowID, sysID].scatter(y=c0_1, x=c0_2, color='black', s=5, alpha=0.8, label=r'Initial Concentration')
        for lineID, aLine in enumerate(TieLineData):
            xAr = np.logspace(np.log10(aLine[2]), np.log10(aLine[3]), 25)
            yAr = pl_line(m=aLine[4], b=aLine[4+2], x=xAr)
            if lineID == 0:
                myAx[rowID, sysID].plot(yAr, xAr, '-', color='orange', alpha=0.8, zorder=0, lw=2, label='Tie-line')
            else:
                myAx[rowID, sysID].plot(yAr, xAr, '-', color='orange', alpha=0.8, zorder=0, lw=2)
    
[anAx.set(xscale='log', yscale='log', xlim=(3e-7, 1.0), ylim=(3e-7, 1.0), 
          xticks=(1e-6, 1e-4, 1e-2), yticks=(1e-6, 1e-4, 1e-2)) for anAx in myAx[-1,:]]
[anAx.set(xlabel='[Whi3 stickers]', ylabel='[$RNA$ stickers]') for anAx in myAx.flatten()]
[anAx.minorticks_off() for anAx in myAx.flatten()]

myAx[0,0].set(xlim=(0.0,0.225), ylim=(0.0, 0.225), xticks=(0.0, 0.1, 0.2), yticks=(0.0, 0.1, 0.2))
myAx[0,1].set(xlim=(0.0,0.225), ylim=(0.0, 0.425), xticks=(0.0, 0.1, 0.2), yticks=(0.0, 0.2, 0.4))
myAx[0,2].set(xlim=(0.0,0.4), ylim=(0.0, 0.525), xticks=(0.0, 0.2, 0.4), yticks=(0.0, 0.2, 0.4))

plHandles, plLabels = plt.gca().get_legend_handles_labels()
myFig.legend(plHandles, plLabels, bbox_to_anchor=(1.0,0.5), loc='center left', fontsize=12, fancybox=False, edgecolor='0.0')

[anAx.text(-0.15, 0.95, f"{plLab}", va='bottom', ha='left', 
           transform=anAx.transAxes, fontweight='bold') for anAx, plLab in zip(myAx.flatten(), 'cdefgh')]

plt.savefig("Fig_S9_V2.pdf", bbox_inches='tight', transparent=True)
plt.show()
```
